```python
# One hot encoding of the alphabets used to represent the amino acids
import numpy
import operator
from functools import reduce
from numpy import argmax
SeqEncoded = []
for j in range(len(allDataDF)):
  data = allDataDF.ModSequence[j]
  # define universe of possible input values
  alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ#/ '
  # define a mapping of chars to integers
  char_to_int = dict((c, i) for i, c in enumerate(alphabet))
  int_to_char = dict((i, c) for i, c in enumerate(alphabet))
  # integer encode input data
  integer_encoded = [char_to_int[char] for char in data]
  #print(integer_encoded)
  # one hot encode
  onehot_encoded = list()
  for value in integer_encoded:
    letter = [0 for _ in range(len(alphabet))]
    letter[value] = 1
    onehot_encoded.append(letter)
  #print(onehot_encoded)
  SeqEncoded.append(onehot_encoded)
# End of the loop
print(len(SeqEncoded))
allDataDF['EncodedX'] = SeqEncoded
```

```python
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

unique_chars = set("".join(allDataDF["ModSequence"]))
vocabulary_size = len(unique_chars)

# Compile models
model1.compile(loss=BinaryCrossentropy(), optimizer=Adam(learning_rate=0.001), metrics=["accuracy"])
model3.compile(loss=BinaryCrossentropy(), optimizer=Adam(learning_rate=0.001), metrics=["accuracy"])

# Training and Timing
start_time_model1 = time.time()
model1.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
end_time_model1 = time.time()

start_time_model3 = time.time()
model3.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
end_time_model3 = time.time()

# Evaluation and Scoring
def evaluate_model(model, X_test, y_test):
  y_pred = model.predict(X_test)
  y_pred_class = y_pred > 0.5  # Assuming threshold 0.5 for binary classification

  accuracy = accuracy_score(y_test, y_pred_class)
  precision = precision_score(y_test, y_pred_class)
  recall = recall_score(y_test, y_pred_class)
  f1 = f1_score(y_test, y_pred_class)
  auc_roc = roc_auc_score(y_test, y_pred)
  return accuracy, precision, recall, f1, auc_roc

# Evaluate models and store results
model_results = {}
model_results["LSTM"] = evaluate_model(model1, X_test, y_test)
model_results["GRU"] = evaluate_model(model3, X_test, y_test)

# Print results in the desired format
print("Model Comparison Results:")
print("| Model | Fitting Time (s) | Scoring Time (s) | Accuracy | Precision | Recall | F1-score | AUC-ROC |")
print("|---|---|---|---|---|---|---|---|")
for model_name, results in model_results.items():
  fitting_time = end_time_model1 - start_time_model1 if model_name == "LSTM" else end_time_model3 - start_time_model3
  accuracy, precision, recall, f1, auc_roc = results
  print(f"| {model_name} | {fitting_time:.4f} | {0.01:.4f} | {accuracy:.4f} | {precision:.4f} | {recall:.4f} | {f1:.4f} | {auc_roc:.4f} |")
```

```python
# Import necessary libraries
import torch
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Load and preprocess the data (assuming you have already done this)
# ...

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(allDataDF['ModSequence'], allDataDF['Value'], test_size=0.2, random_state=42)

# DistilBERT
# Load the pre-trained DistilBERT model and tokenizer
distilbert_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
distilbert_model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased')

# Tokenize the input sequences
X_train_tokens = distilbert_tokenizer(list(X_train), padding=True, truncation=True, max_length=22, return_tensors='pt')
X_test_tokens = distilbert_tokenizer(list(X_test), padding=True, truncation=True, max_length=22, return_tensors='pt')

# Set the device (GPU if available, else CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move the model and data to the device
distilbert_model = distilbert_model.to(device)
X_train_tokens = {k: v.to(device) for k, v in X_train_tokens.items()}
y_train = torch.tensor(y_train, dtype=torch.long).to(device)

# Define the optimizer and loss function
optimizer = torch.optim.Adam(distilbert_model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
# Training loop
epochs = 1
for epoch in range(epochs):
    distilbert_model.train()
    total_loss = 0
    for i in tqdm(range(len(X_train_tokens['input_ids'])), desc=f"Epoch {epoch + 1}"):
        optimizer.zero_grad()
        output = distilbert_model(input_ids=X_train_tokens['input_ids'][i].unsqueeze(0),
                                   attention_mask=X_train_tokens['attention_mask'][i].unsqueeze(0),
                                   labels=y_train[i].unsqueeze(0))
        loss = output.loss
        total_loss += loss.item()
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch + 1} loss: {total_loss / len(X_train_tokens['input_ids'])}")

```

```python
# Evaluate the DistilBERT model
distilbert_model.eval()
with torch.no_grad():
    y_pred_distilbert = distilbert_model(input_ids=X_test_tokens['input_ids'], attention_mask=X_test_tokens['attention_mask'])
    y_pred_distilbert = y_pred_distilbert.logits.argmax(dim=-1).detach().cpu().numpy()

distilbert_accuracy = accuracy_score(y_test, y_pred_distilbert)
distilbert_precision = precision_score(y_test, y_pred_distilbert)
distilbert_recall = recall_score(y_test, y_pred_distilbert)
distilbert_f1 = f1_score(y_test, y_pred_distilbert)
distilbert_roc = roc_auc_score(y_test, y_pred_distilbert)

```

```python
# RoBERTa
# Load the pre-trained RoBERTa model and tokenizer
roberta_tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
roberta_model = RobertaForSequenceClassification.from_pretrained('roberta-base')

# Tokenize the input sequences
X_train_tokens = roberta_tokenizer(list(X_train), padding=True, truncation=True, max_length=22, return_tensors='pt')
X_test_tokens = roberta_tokenizer(list(X_test), padding=True, truncation=True, max_length=22, return_tensors='pt')

# Move the model and data to the device
roberta_model = roberta_model.to(device)
X_train_tokens = {k: v.to(device) for k, v in X_train_tokens.items()}
y_train = torch.tensor(y_train, dtype=torch.long).to(device)

# Define the optimizer and loss function
optimizer = torch.optim.Adam(roberta_model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# Initialize the optimizer state
for param_group in optimizer.param_groups:
    for param in param_group['params']:
        if param.requires_grad:
            param.data.zero_()
            if param.dim() > 1:
                param.grad = torch.zeros_like(param.data)
            else:
                param.grad = param.data.new().resize_(*param.data.size())
```

```python
# Training loop
epochs = 1
for epoch in range(epochs):
    roberta_model.train()
    total_loss = 0
    for i in tqdm(range(len(X_train_tokens['input_ids'])), desc=f"Epoch {epoch + 1}"):
        optimizer.zero_grad()
        output = roberta_model(input_ids=X_train_tokens['input_ids'][i].unsqueeze(0),
                                attention_mask=X_train_tokens['attention_mask'][i].unsqueeze(0),
                                labels=y_train[i].unsqueeze(0))
        loss = output.loss
        total_loss += loss.item()
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch + 1} loss: {total_loss / len(X_train_tokens['input_ids'])}")
```

```python
# Evaluate the RoBERTa model
roberta_model.eval()
with torch.no_grad():
    y_pred_roberta = roberta_model(input_ids=X_test_tokens['input_ids'], attention_mask=X_test_tokens['attention_mask'])
    y_pred_roberta = y_pred_roberta.logits.argmax(dim=-1).detach().cpu().numpy()

roberta_accuracy = accuracy_score(y_test, y_pred_roberta)
roberta_precision = precision_score(y_test, y_pred_roberta)
roberta_recall = recall_score(y_test, y_pred_roberta)
roberta_f1 = f1_score(y_test, y_pred_roberta)
roberta_roc = roc_auc_score(y_test, y_pred_roberta)

# Create a DataFrame to store the results
transformer_models = pd.DataFrame({
    'Model': ['DistilBERT', 'RoBERTa'],
    'Accuracy': [distilbert_accuracy, roberta_accuracy],
    'Precision': [distilbert_precision, roberta_precision],
    'Recall': [distilbert_recall, roberta_recall],
    'F1_score': [distilbert_f1, roberta_f1],
    'AUC_ROC': [distilbert_roc, roberta_roc]
})
```
